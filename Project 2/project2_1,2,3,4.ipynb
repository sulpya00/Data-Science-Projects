```python
#Rows with duplicate data
duplicate_rows_iris = iris[iris.duplicated()]
print("number of duplicate rows: ", duplicate_rows_iris.shape)
```

```python
#Drop duplicate
iris=iris.drop_duplicates()
```

```python
#Sepal Length
mean_data = iris["sepal_length"].mean()
median_data = iris["sepal_length"].median()
min_data=iris["sepal_length"].min()
max_data=iris["sepal_length"].max()
  
print("Mean:", mean_data, "\nMedian:",median_data, "\nMinimum:",min_data, "\nMaximum:", max_data)
```

```python
#Petal Length
mean_data = iris["petal_length"].mean()
median_data = iris["petal_length"].median()
min_data=iris["petal_length"].min()
max_data=iris["petal_length"].max()
  
print("Mean:", mean_data, "\nMedian:",median_data, "\nMinimum:",min_data, "\nMaximum:", max_data)
```
